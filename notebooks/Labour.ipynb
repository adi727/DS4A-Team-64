```python
df = pd.read_table('metro.txt', sep='|', names= ('LAUS_AREA_CODE','FIPS_CD_ST','FIPS_CD_CT','AREA_TITLE','PERIOD','LABOUR_FORCE','EMPLOYED','UNEMP_LEVEL','UNEMP_RATE'))
```

|FIPS CD CT     |Federal Information Processing Standard code for state|

```python
df.dropna(inplace = True)
```

```python
df.isna().sum()
```

##### Split AREA_TITLE column to county and state

```python
df[['COUNTY','STATE']] =  df.AREA_TITLE.str.split(',',expand=True,)
df
```

##### Split the PERIOD column into PERIODC and DELETE, with plans to remove (p)

```python
df[['PERIODC','DELETE']]= df['PERIOD'].astype(str).str.split('(',expand = True,)
df
```

##### Remove the DELETE column 

```python
del df['DELETE']
```

##### Swap columns PERIOD and PERIODC

```python
df['PERIOD'],df['PERIODC'] = df['PERIODC'],df['PERIOD']
df
```

##### Delete column PERIODC 

```python
del df['PERIODC']
```

##### convert column PERIOD to datetime object

```python
df['PERIOD']= pd.to_datetime(df['PERIOD'].str.strip(), format = '%b-%y')
df
```
